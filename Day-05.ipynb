```python
score = int(input("Enter score : "))

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print("Grade :", grade)
```

```python
# check even or odd

num = int(input("Enter a number : "))

if num % 2 == 0:
    print(num, "is an even number.")
else:
    print(num, "is an odd number.")
```

```python
# check num is multiple of 3 & 5

num = int(input("Enter a number : "))

if num % 3 == 0 and num % 5 == 0:
    print(num, "is a multiple of both 3 and 5.")
else:
    print(num, "is not a multiple of both 3 and 5.")
```

```python
li = [1,2,3,4,5,6,7,8,9]
li.pop()
print(li)
li1 =[]
sum = 0
for i in li:
    if i % 2 == 0:
        sum+=i
        li1.append(i)
    else :
        continue

print("Sum of even numbers in list is :", sum)
print("New List :", li1)
```

```python
li = [10, 20, 30, 40, 50]
print(li)

# list Methods
li.append('abc')
print(li)

li2 = [60, 90, 80]

li.extend(li2)
print(li)

li.pop()
print(li)

li.remove(30)
print(li)

li.insert(2, 25)
print(li)

li2.sort()
print(li2)

li.reverse()
print(li)

print(li.index(20))  

print(li.count(10))
```

```python
# Leap year
year = int(input("Enter a year : "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
bread = ["Rye Bread", "White", "Wheat"]
meat = ['Meayball', 'Sausages','Chicken']
veg = ['Lettuce', 'Tomato', 'cucumber']
sauses = ['mayo', 'mustard', 'chilli']
count = 0
for b in bread:
    for m in meat:
        for v in veg:
            for s in sauses:
              print(f"{b} + {m} + {v} + {s}")
              count += 1
print(count)
```

```python
# Factorial

num = int(input("Enter a number : "))
fact = 1

for i in range(1, num+1):
    fact *= i
print(fact)
```
