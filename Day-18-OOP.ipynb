# Object-Oriented Programming (OOP)

## 1. Core Concepts
* **Object:** A real-world entity that has **State** (Properties/Attributes) and **Behavior** (Functionality/Methods).
* **Class:** The **blueprint** or template used to create objects. It defines the structure (properties) and behavior that the objects will have.

## 2. Function vs Method
* **Function:** A block of code defined **outside** a class. It is called directly by its name.
    * *Call:* `function_name()`
* **Method:** A function defined **inside** a class. It effectively belongs to an object and is called using that object.
    * *Call:* `object.method_name()`

---

## 3. Memory Management (Important)
Understanding where data lives is key for performance and debugging.

* **Heap Memory:** Stores the **actual Object** (the data).
* **Stack Memory:** Stores the **Reference** (the variable name/pointer) that points to the object in the Heap.
* **The Flow:** When you access an object, the program looks at the **Stack** to find the address, then goes to the **Heap** to retrieve the data.



> **Key Difference:**
> * **Class:** Does **not** occupy memory for data (it's just a definition/code).
> * **Object:** Occupies actual memory space when instantiated.

## The `self` Parameter

In Python classes, methods must have an extra first parameter in their definition. By convention, this is named `self`.

* **Purpose:** It acts as a reference to the **current object** (instance) calling the method.
* **Access:** It allows the code to access variables (attributes) and other methods that belong to that specific object.
* **Automatic Passing:**
    * **Definition:** You *must* include `self` when defining the method.
    * **Calling:** You *do not* pass `self` when calling the method; Python passes it automatically.

````python
class car:
    color = "red"
    brand = "Toyota"

    def start(self):
        print("Car started")

    def stop(self):
        print("Car stopped")
        
c = car()
c.start()
c.stop()

print(type(c))
````

# Instance Variables (`self.variable`)

Using `self.variable_name` attaches data to the specific **object instance**. This allows the data to be accessed by **any method** within that class.

### The Critical Difference
1.  **`variable = 10` (Local Variable):**
    * Temporary.
    * Only exists inside the function where it is created.
    * Cannot be accessed by other methods.

2.  **`self.variable = 10` (Instance Variable):**
    * Permanent (as long as the object exists).
    * Stored in the object's memory (Heap).
    * Can be accessed by **any** method in the class using `self`.

````python
class student:
    name = "John"
    contact = 1234567890
    email = 'abc@gmail.com'

    def display(self, name=name, contact=contact, email=email):
        self.name = name
        self.contact = contact
        self.email = email
        print("Name:", self.name)
        print("Contact:",self.contact)
        print("Email:", self.email)

s = student()
s.display('prathamesh', 12346, 'pra@gmail.com')
````

## Types of Methods in Python

There are distinct types of methods in Python classes, distinguished by how they access data (State).

### 1. Instance Method
The most common type of method. It acts upon a specific object (instance) of the class.
* **Behavior:** Can access and modify unique data of the specific object (`self.variable`).
* **Parameter:** Must have `self` as the first parameter.
* **Decorator:** None required (default behavior).

### 2. Class Method
A method that is bound to the **Class**, not the object. It can modify the class state that applies across all instances.
* **Behavior:** Can access class variables but **cannot** access specific instance variables (`self`).
* **Parameter:** Must have `cls` as the first parameter (stands for Class).
* **Decorator:** Requires `@classmethod`.

### 3. Static Method
A utility method that belongs to the class namespace but **does not** access or modify the class or instance state. It is just a regular function that happens to live inside a class.
* **Behavior:** Cannot access `self` or `cls`. It is isolated.
* **Parameter:** No special first parameter (`self`/`cls` are not passed).
* **Decorator:** Requires `@staticmethod`.

### 4. Special Methods (Magic/Dunder Methods)
These methods start and end with double underscores (`__`). They are not called directly but are invoked automatically by Python operators or built-in functions.

* **`__init__(self)` (Constructor):**
    * **Trigger:** Automatically called when a new object is created.
    * **Purpose:** Used to initialize the object's attributes (variables).

* **`__str__(self)` (String Representation):**
    * **Trigger:** Automatically called when you use `print(object)` or `str(object)`.
    * **Purpose:** Returns a human-readable string description of the object. If missing, Python prints the confusing memory address (e.g., `<__main__.Obj object at 0x00...>`).



### Summary Comparison Table

| Method Type | Decorator | First Arg | Can Modify Object State? | Can Modify Class State? |
| :--- | :--- | :--- | :---: | :---: |
| **Instance** | None | `self` | âœ… Yes | âœ… Yes |
| **Class** | `@classmethod` | `cls` | âŒ No | âœ… Yes |
| **Static** | `@staticmethod` | None | âŒ No | âŒ No |

````python
class product():
    product = "mobile"
    
    @classmethod
    def show(cls):
        print("Product:", cls.product)

p = product()
p.show()
product.show()		# nameless object -> used in inheritence
````

````python
# static method

class math: 
    @staticmethod
    def add(a, b):
        return a + b

# print(math.add(5, 10))
m = math()
m.add(5, 10)
````

````python
# constructor and __str__ method

class employee:
    def __init__(self, eid, name, salary):
        self.eid = eid
        self.name = name
        self.salary = salary
    def __str__(self):
        return f"\nEmployee ID: {self.eid}, \nName :{self.name}, \nSalary :{self.salary}"
        
e = employee(101, "qwerty", 456789)
print(e)

print(employee(102, "asdfgh", 123456))
````

# Encapsulation

Encapsulation is the process of wrapping data (variables) and methods together into a single unit (Class). It also **restricts direct access** to some of an object's **internal details** to prevent accidental modification.

## Access Modifiers
Python controls access using **naming conventions** (underscores) rather than keywords like `public` or `private` used in Java/C++.

### 1. Public Member
* **Symbol:** No underscores (e.g., `self.name`).
* **Access:** Can be accessed from **anywhere** (inside or outside the class).
* **Default:** All variables/methods are public by default in Python.

### 2. Protected Member
* **Symbol:** Single underscore prefix (e.g., `self._name`).
* **Access:** Intended to be accessed only within the **class** and its **subclasses** (Inheritance).
* **Note:** This is a convention. Python does not strictly enforce it, but developers respect it as a warning: *"Do not touch this from outside."*

### 3. Private Member
* **Symbol:** Double underscore prefix (e.g., `self.__name`).
* **Access:** Can **only** be accessed within the **class itself**.
* **Mechanism:** Python performs **Name Mangling** (changes the name internally to `_ClassName__variable`) to make it hard to access from outside.



### Summary Table

| Modifier | Syntax | Scope |
| :--- | :--- | :--- |
| **Public** | `name` | Accessible Everywhere |
| **Protected** | `_name` | Class + Subclasses |
| **Private** | `__name` | Class Only |

````python
# public

class laptop:
    def __init__(self, laptop):
        self.laptop = laptop
    def show(self):
        print("Laptop:", self.laptop)
l = laptop("Lenovo")
l.show()
````

````python
# protected

class laptop:
    def __init__(self, laptop):
        self._laptop = laptop
    def show(self):
        print("Laptop:", self._laptop)
l = laptop("Lenovo")
l.show()
print("Protected member:", l._laptop)  # accessible but not recommended
````

````python
# private

class laptop:
    def __init__(self, laptop):
        self.__laptop = laptop
    def show(self):
        print("Laptop:", self.__laptop)
l = laptop("Lenovo")
l.show()
# print("Private Member: ", l.__laptop)		# will give error
````

# Inheritance (IS-A relationship)

Inheritance is a mechanism where a new class (Child/Derived Class) acquires the properties and behaviors (methods) of an existing class (Parent/Base Class). This promotes **code reusability**.

## Types of Inheritance

### 1. Single Inheritance
One child class inherits from **one** parent class.
* **Flow:** A â†’ B

### 2. Multiple Inheritance
One child class inherits from **two or more** parent classes.
* **Flow:** A + B â†’ C
* *Note:* Python supports this natively (unlike Java).

### 3. Multi-level Inheritance
A child class inherits from a parent, which in turn inherits from another parent (a chain).
* **Flow:** A â†’ B â†’ C (C inherits from B, B inherits from A).

### 4. Hierarchical Inheritance
**Multiple** child classes inherit from a **single** parent class.
* **Flow:** A â†’ B and A â†’ C.

### 5. Hybrid Inheritance
A combination of two or more types of inheritance (e.g., Hierarchical + Multiple).
* **Diamond Problem:** This often leads to complexity (the "Diamond Problem"), which Python resolves using MRO (Method Resolution Order).

````python
# multiple inheritance

class a:
    fname = "John"

class b:
    lname = "Doe"
    fname = "Doe"
    
class c(a, b):
    def show(self):
        print("Full Name:", self.fname, self.lname)
        print("Full Name:", self.fname, self.fname)
        print("Full Name:", self.fname, b.fname)

m = c()
m.show()
````

# Polymorphism
The word "Polymorphism" means **"many forms."** In programming, it allows the same function, operator, or object to behave differently depending on the context.

## 1. Polymorphism with Functions and Objects
Python built-in functions can work on different data types.
* **Example:** The `len()` function.
    * `len("Hello")`: Returns **5** (counts characters).
    * `len([1, 2, 3])`: Returns **3** (counts elements).
    * *Same function name, different behavior based on the input type.*

## 2. Polymorphism with Operators (Operator Overloading)
The same operator symbol can perform different actions based on the operands.
* **Example:** The `+` symbol.
    * `4 + 5` = `9` (Addition).
    * `"Python" + "Code"` = `"PythonCode"` (Concatenation).

## 3. Polymorphism with Class Methods
Different classes can have methods with the **same name**. This allows you to loop through a list of different objects and call the same method on each one without worrying about their specific class.

## 4. Duck Typing
> **The Philosophy:** "If it looks like a duck, swims like a duck, and quacks like a duck, then it probably is a duck."

In Python, we do **not** check the class type of an object (e.g., `isinstance(obj, Duck)`). Instead, we check if the object **has the specific method** or attribute we need.

* **Dynamic Behavior:** Python assumes that if an object has the required method, it can be used, regardless of its class hierarchy.



````python
# polymorphism
# same functions working with different data types

# 1. with function and objects
print(len([1,2,3,4]))
print(len("Hello"))
print(len({1: 'a', 2: 'b',}))

# 2. with operators
print([1,2,3,4]*4)
````

````python
# with class methods (method Overriding)
# different class methods with same name but different behavior

class dog:
    def sound(self):
        print("Bark")
        
class cat:
    def sound(self):
        print("Meow")
        
for animal in [dog(), cat()]:
    animal.sound()
````

````python
# with duck typing
# does'nt care about the type of object, if it has the required method, it can be used

class bike:
    def run(self):
        print("Bike is running")
        
class boat:
    def run(self):
        print("boat is running")
        
def transport(vehcile):
    vehcile.run()

transport(bike())
transport(boat())
````

# Composition (HAS-A Relationship)

Composition is a design principle where a class **contains** an object of another class as an instance variable, rather than inheriting from it.

* **Inheritance (IS-A):** "A Dog **is a** Animal."
* **Composition (HAS-A):** "A Car **has an** Engine."

## How it Works
Instead of inheriting all functionality, a class delegates some responsibilities to other objects. If the container object (e.g., Car) is destroyed, the contained objects (e.g., Engine) are effectively destroyed with it in this context.

````python
class engine:
    def start(self):
        print("engine startedðŸš‚")

class wheel:
    def rotate(self):
        print("Wheel Rotating ðŸš‚ðŸ›žðŸ›ž")
        
class car:
    def __init__(self):
        self.engine = engine()
        self.wheel = [wheel() for _ in range(4)]
        
    def drive(self):
        print(self.engine.start())
        for wheel in self.wheel:
            print(wheel.rotate())
            
c = car()
print(c.drive())
````

# Abstraction
Abstraction focuses on **hiding unnecessary implementation details** and **showing only the essential feature**s of an object. It allows the user to know *what* an object does without needing to know *how* it does it.

* **Real-World Example:** You know how to drive a car (steering, pedals) without understanding the complex fuel injection logic inside the engine.

## Key Components in Python

### 1. Abstract Base Class (ABC)
A class that **cannot be instantiated** directly. It serves as a blueprint for other classes.
* **Purpose:** It forces all child classes to follow a specific structure.
* **Module:** Requires importing `ABC` from the `abc` module.

### 2. Abstract Method
A method that is declared in the Abstract Base Class but contains **no implementation** (pass).
* **Rule:** Any child class inheriting from the ABC **must** override and implement this method. If it fails to do so, Python will raise an error.
* **Decorator:** Requires `@abstractmethod`.

````python
from abc import ABC, abstractmethod

class fuelError:
    pass

# composition
class engine:
    def __init__(self, horsepower, fuel):
        self.horsepower = horsepower
        self.fuel = fuel
    def start(self):
        return f'engine started with {self.horsepower} HP using {self.fuel}'

# abstraction
class auto(ABC):
    vehicle_count = 0
    def __init__(self, brand, model, price):
        self._brand = brand
        self._model = model
        self._price = price
        auto.vehicle_count += 1

    @abstractmethod
    def mileage(self):
        pass
    
    def get_details(self):
        return f'{self._brand}, {self._model} , â‚¹{self._price}'
    
    def vehicle_type(self):
        pass

    # using class method
    @classmethod
    def total_vehicle(cls): 
        return cls.vehicle_count

    #static method
    @staticmethod
    def company_policy():
        return "all cars runs"

class car(auto):
    def __init__(self, brand, model, price, engine, fuel):
        super().__init__(brand, model, price)
        self.engine = engine  # HAS-A Relationship
        self.fuel = fuel
        
    def mileage(self):
        return '15km/L'

    def vehicle_type(self):
        return "car"
    
    def refuel(self, liters):
        if liters <= 0:
            raise fuelError("invalid fuel amount")
        self.fuel += liters
        return f"refueled {liters} liters"

class ElectricCar(car):
    def mileage(self):      # polymorphism
        return "120km/kw"

    def vehicle_type(self):
        return "EV"

    def recharge(self, charge):
        if charge <= 0:
            raise fuelError("invalid charge amount")
        self.fuel += charge
        return f"recharged {charge} kW"

# operator overloading
class showroom:
    def __init__(self, name):
        self.name = name
        self.cars = []

    def addCar(self, other):
        self.cars.append(car)
    
    def __add__(self, other):
        newShowroom = showroom(self.name + " & " + other.name)
        return newShowroom
    
    def display(self):
        return [car.get_details() for car in self.cars]
    
    # polymorphuim function
    def printVehicleInfo(self, vehicle):
        print(vehicle.get_details())
        print("Type:", vehicle.vehicle_type())
        print("Mileage : ", vehicle.mileage())

# main program
if __name__ == "__main__":
    try:
        engine1 = engine(120, "Petrol")
        engine2 = engine(150, "Electric")

        car1 = car("Toyota", "Corolia", 150000, engine1, 40)
        car2 = ElectricCar("Tesla", "Model X", 600000,engine2, 0)

        engine1.start()
        engine2.start()


        showroom1 = showroom("City Motors")
        showroom2 = showroom("qwerty Motors")

        showroom1.printVehicleInfo(car1)
        showroom2.printVehicleInfo(car2)

        showroom1.addCar(car1)
        showroom2.addCar(car2)

        mega_showroom = showroom1 + showroom2

        for car in mega_showroom.display():
            print(car)

        print("Total Vehicle : ", auto.total_vehicle())
        print("Policy : ", auto.company_policy())

        print(car1.refuel(10))
        print(car2.recharge(20))

    except fuelError as e:
        print("Fuel Error: ", e)
````

````python
from abc import ABC, abstractmethod

# abstract class
class payment(ABC):
    
    @abstractmethod
    def pay(self, amount):
        pass

class creditCard(payment):
    def pay(self, amount):
        print(f"Paid {amount}")
    
class paypal(payment):
    def pay(self, amount):
        print(f"Paid {amount}")

def make_payment(payment_method, amount):
    payment_method.pay(amount)

cd = creditCard()
ppal = paypal()

make_payment(cd, 1000)
make_payment(ppal, 200)
````
