## 1. Read and save data from file 

- `r`: Open for reading only. File must exist or raises FileNotFoundError.
- `w`: Open for writing; truncates existing file or creates new one.
- `x`: Open for exclusive creation; fails if file already exists immediately.
- `a`: Open for appending; writes go to file end and create.
- `b`: Binary mode for file operations; reading and writing use bytes.
- `t`: Text mode (default); reads and writes expect strings, universal newline.
- `+`: Updating mode; opens for reading and writing, doesn't truncate file.
- `wt` : Open file in write text mode

- `os.remove` : to remove a file

```python
# 2. Dictionary Methods

str1 = ['a', 'b', 'c', 'a', 'b', 'c']
dict1 = {}

for ch in str1:
    if ch not in dict1.keys():
        dict1[ch] = 0
    dict1[ch] += 1
print(dict1)
```

```python
str1 = ['a', 'b', 'c', 'a', 'b', 'c']
dict1 = {}

for ch in str1:
    dict1.setdefault(ch, 0)
    dict1[ch] += 1
    print(list(dict1.items()))
dict1
```

```python
dic = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
dic.setdefault('e', 5)
print(dic)

dic.update({'e': 6 })
dic.update({'k': 6 })
dic.update(h= 7)
print(dic)
```
