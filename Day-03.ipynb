```python
# User input and Addition

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("{} + {} = {}".format(a, b, a + b))
print("{} - {} = {}".format(a, b, a - b))
print("{} * {} = {}".format(a, b, a * b))
print("{} / {} = {}".format(a, b, a / b))
print("{} % {} = {}".format(a, b, a % b))
print("{}**2,  {}**2 = {}, {}".format(a, b, a**2, b**2))
```

```python
# String OperTIONS

s = "hello"

print(s.upper())
print(s.lower())
print(s.capitalize())

print(s.startswith('h'))
print(s.endswith('o'))

print(len(s))
print(s.__len__())

print(s.find('l'))
```

```python
# indexing and Slicing

a = 'abcdefghijklmnopqrstuvwxyz'


print(a[0])
print(a[1])
print(a[-1])
print(a[::-1])

print(a[2:7])
# print(a[-2:-7])

p = 'pcuuniversity'

print(p[-2:-9:-1])    # [starting_index, ending_index - 1, step_index (if wanted in reverse '-1')]
print(p[0:15:2])
```
