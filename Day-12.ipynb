## Set Operations

1. union
2. Difference
	- 2.1 symmetric Difference

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 | s2		#union
```

```python
s1 & s2		# intersection
s1.union(s2)
```

```python
s1 - s2 	# difference
s1.difference(s2)
```

```python
s1 ^ s2 	#symmetric diff
```
