## 1. List and Tuple

```python
p1 = ['David', 20, 1, 180.0, 100.0]
p2 = ['john', 25, 1, 170.0, 80.0]
p3 = ['jane', 22, 0, 169.0, 60.0]
p4 = ['peter', 40, 1, 150.0, 50.0]

p_list = p1 + p2 + p3 + p4

n_person = int(len(p_list) / 5)
age_sum = 0.0
for age in p_list[1::5]:
    age_sum += age
age_ave = float(age_sum / n_person)
print("Average age: ", str(age_ave))
```

## 2. Dictionary

```python
d = {'Name' : 'Prathamesh', 
     'Age' : 20, 
     'Div' : 'B'}

print(d)

print('Name : ',d['Name'])
print('Age : ',d['Age'])
print('Div : ',d['Div'])


# Dictionary Metheds
d['Roll'] = 101 
print(d)

print(d.keys())

print(d.get('Name'))

print(d.values())

print(d.pop('Roll'))
```
