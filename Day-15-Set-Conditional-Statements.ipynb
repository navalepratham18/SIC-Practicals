# Set

## 1. What is a Set?
A Set is a collection of **unique elements** stored in an **unordered** format. It is defined using curly braces `{}`.

### Characteristics:
1.  **Unordered (No Indexing):**
    * Items do not have a defined order.
    * You **cannot** access elements using an index (e.g., `my_set[0]` will throw an error).
2.  **Stores Unique Elements:**
    * Duplicate values are automatically removed.
    * Useful for removing duplicates from a list.
3.  **Mutable:**
    * You can add or remove elements from the set after creation.
    * *Note: The elements inside the set must be immutable (hashable).*
4.  **Fast Performance:**
    * Sets use a **Hash Table** structure internally.
    * Checking if an item exists (`x in set`) is extremely fast (O(1) average time complexity), unlike lists which require a linear search (O(n)).

---

## 2. Set Operations
Since sets follow mathematical set theory, they support the following operations:

### Common Set Methods (Modification)

1. **`add(element)`**
   * Adds a single element to the set.
   * If the element already exists, it does nothing (no error).

2. **`pop()`**
   * Removes and returns an **arbitrary (random)** element from the set.
   * *Warning:* Since sets are unordered, you cannot predict which item will be removed. Raises an error if the set is empty.

3. **`remove(element)`**
   * Removes a specific element.
   * **Crucial:** If the element is **not found**, it raises a `KeyError`.

4. **`discard(element)`**
   * Removes a specific element.
   * **Crucial:** If the element is **not found**, it does **nothing** (no error is raised). Safer to use than `remove()` if you are unsure the item exists.


| Operation | Operator | Method | Description |
| :--- | :---: | :--- | :--- |
| **Union** | `|` | `setA.union(setB)` | Combines all elements from both sets (removes duplicates). |
| **Intersection** | `&` | `setA.intersection(setB)` | Returns only elements found in **both** sets. |
| **Difference** | `-` | `setA.difference(setB)` | Returns elements in A that are **not** in B. |
| **Symmetric Difference** | `^` | `setA.symmetric_difference(setB)` | Returns elements in A or B, but **not both**. |

# Conditional Statements

Conditional statements control the flow of execution. They allow the program to **execute different blocks of code** based on whether a specific condition evaluates to `True` or `False`.

## Types of Conditional Statements

* **`if`:** Executes a block of code strictly when the specified condition is True.
* **`elif`:** Checks an alternative condition only if the preceding `if` (or `elif`) was False.
* **`if-else`:** Executes the main block if the condition is True, otherwise automatically runs the backup `else` block.

```python
# Nested if 

num = [1,2,3,4,5]

if 1 in num:
	print("1")
if 2 in num:
	print("2")
if 4.5 in num:
	print("4.5")
```

```python
# if-elif-else 

mark = int(input("Enter your mark: "))
if mark >= 90:
    print("grade A")
elif mark >= 75:
	print("grade B")
elif mark < 75 and mark >= 50:
     print("grade C")
else:
    print("grade D")
```

```python
# Nested if to check positive or negative

num = int(input("Enter a number: "))

if num != 0:
    if num >= 0:
        print(f"{num} Positive number")
    else:
        print(f"{num} Negative number")
else : 
    print("Zero")
```
