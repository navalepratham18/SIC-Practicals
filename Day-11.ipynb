## 1. Dictionary Methods

- `pop()` : Removes a specific key-value pair
- `popitem()` : REmoves last key-value pair
- `clear()` : Clear the entire dictionary
- `copy()` : copies the entire dictionary the another dictioanry
- `key()` : REturns all the keys from dictionary 
- `values()` : Returns all the values from dictionary
- `get(key)` : returns the value at key 
- `items()` :  Returns dictionary

---

## 2. Text Data

- `split()` : Seperates the string and stores in list
- `strip()` : Delete spaces before and after seperator character
  - `lstrip()` : Deletes spaces from left side
  - `rstrip()` : deletes spaces from right side
- `.join()` : connecting strings
- `.replace()` : Replaces the specified element  

---

## 3. Dictionary Conversion

- `fromkeys(key, value)` : Assign values to dictionary if keys are specified
- `defaultdict( datatypes )` : if a key is absent in, then it automatically defines it as 'int 0'

## 3. Dictionary conversion

```python
keys = ['a', 'b', 'c']      #list
x = dict.fromkeys(keys)
x
```

```python
y = dict.fromkeys(keys, 100)
y
```

```python
from collections import defaultdict

keys = ('a', 'b', 'c', 'd')
y = dict.fromkeys(keys, 100)
y = defaultdict(int)

y['z']
# y
```

```python
y
```

```python
from collections import defaultdict

word = 'prathamesh'
counter = defaultdict(int)
for l in word : 
    counter[l] +=1

counter
```

```python
x = {'a' : 10, 'b' : 20, 'c' : 30, 'd' : 40}

for key in x.keys():
    print(key, end=' ')
```

```python
x = {'a' : 10, 'b' : 20, 'c' : 30, 'd' : 40}

for key in x.values():
    print(key, end=' ')
```
