## JSON Data Dictionary / list Conversion

### 1. json

```python
import json

data = '{"Name": "David", "Age": 25, "HObhy": ["basketball"],"family": {"Father": "John", "mother": "marry"}, "Married": true}'

json_data = json.loads(data)

json_data1 = json.dumps(json_data)

print(type(json_data))
print(json_data['Name'])


print(json_data1)
data
```

```python
data = '{"title": "the old man and the Sea", "ISBN": "12345", "Author": "Ernest Hemingway"}'

json_data = json.loads(data)
print(json_data)

with open('book.json', 'w') as f:
    json.dump(json_data, f)
```

```python
list1 = [1, 2, 3, 4, 5]

print("befor pop(0)", list1)
print("before pop(1)", list1[1])
list1.pop(0)
print("after pop(0)", list1)
print("after pop(1)", list1[1])
```

```python
dict1 = {0:11, 1:22, 2:33, 3:44, 4:55}

print("before pop(0): ", dict1.items())
print("before pop(1): ", dict1[1])
dict1.pop(0)
print("after pop(0): ", dict1.items())
print("after pop(1): ", dict1[1])
```
