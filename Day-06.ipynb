```python
i = 1
while i<11:
   print(i*2, end=', ')
   i+=1
```

```python
i=0
while i < 10:
    if i==5:
       print(i)
       break
    i+=1
```

```python
# continue statement
i=0
while i < 10:
    i+=1
    if i==5:
       continue
    print(i)
```

```python
n = int(input("Enter a number: "))
s= 0
i=1
while i <= n:
    s += i
    i += 1
print("Sum of 1 to {} is {}".format(n,s))
```

```python
selected = None
while selected not in ['Scissor','Rock', 'Paper']:
    selected = input("Choose among Scissor, Rock, Paper: ")
    print("You selected: ", selected)
```

```python
n = int(input("Enter a number: "))
s = 0
for i in range(1,n+1):
    s += i
print("Sum of 1 to {} is {}".format(n,s))


# in python range function; it always works on the positive values 
```
